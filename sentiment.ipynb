```python
from textblob import TextBlob
import ipywidgets as widgets
from IPython.display import display, HTML

# Create Text Input Widget
text_input = widgets.Textarea(
    value='',
    placeholder='Type your text here...',
    description='Text:',
    layout=widgets.Layout(width='100%', height='120px'),
    style={'description_width': 'initial'}
)

# Create Output Area
output = widgets.Output()

# Analyze Function
def analyze_sentiment(change):
    output.clear_output()
    text = text_input.value.strip()

    if not text:
        with output:
            display(HTML('<p style="color:orange;">Please enter some text for analysis.</p>'))
        return

    blob = TextBlob(text)
    polarity = blob.sentiment.polarity
    subjectivity = blob.sentiment.subjectivity

    if polarity > 0:
        sentiment = "Positive"
        color = "green"
    elif polarity < 0:
        sentiment = "Negative"
        color = "red"
    else:
        sentiment = "Neutral"
        color = "blue"

    result_html = f"""
    <p><b>Sentiment:</b> <span style="color:{color};">{sentiment}</span></p>
    <p><b>Polarity:</b> {polarity:.2f}</p>
    <p><b>Subjectivity:</b> {subjectivity:.2f}</p>
    """

    with output:
        display(HTML(result_html))

# Create Analyze Button
analyze_button = widgets.Button(
    description='Analyze Sentiment',
    button_style='success',
    layout=widgets.Layout(width='30%'),
    tooltip='Click to analyze sentiment'
)
analyze_button.on_click(analyze_sentiment)

# Display the UI
display(widgets.Label("🧠 Sentiment Analyzer (using TextBlob)"))
display(text_input, analyze_button, output)
```
